```python
from openai import OpenAI

client = OpenAI(api_key='pplx-', base_url="https://api.perplexity.ai")

# Define a function to generate the structured prompt for each chapter
def generate_prompt(row):
    return f"""Create a Captivating story for Chapter {row['Chapter']}:{row['Title']}  with below format. below is the sample format. make sure that the story is authentic and natural and only use the concepts from the chapter outline. do not add unnecessary headings or sub headings
Chapter Title: {row['Title']}
Engaging Introduction (Paragraph 1):
Craft an opening that captures the reader's attention by reintroducing Christopher Columbus and setting the stage for the next part of his adventure. Ensure continuity by referencing key events or decisions from previous chapters, highlighting how they have led to this moment. For example, Columbus has been preparing for his voyage for months, gathering supplies and assembling a diverse crew of 90 sailors. He is determined to reach the Caribbean within 30 days, navigating through unpredictable ocean currents that can affect his speed by up to 5 km/h.
The Vision/Proposal/Challenge (Paragraph 2):
Describe Columbus's vision or the specific challenge he faces in this chapter, ensuring it aligns with the chapter's themes and topics. This should build upon the progress made in previous chapters, showcasing how Columbus's journey is evolving. For instance, Columbus aims to explore new landforms and encounter indigenous cultures, which will require him to be resilient in the face of uncertainty. He must use his knowledge of astronomy to navigate, relying on constellations like the North Star to guide him.
Integration of Key Topics (Paragraph 3):
Incorporate all relevant topics from the chapter outline naturally and subtly, while preserving the following concepts:
Math Focus:{row['Math Focus']} - For example, Columbus uses geometry to pack supplies efficiently, ensuring that his ships can carry enough food for the entire crew. He calculates that each sailor needs about 2 kg of food per day, which requires careful budgeting.
Science Focus: {row['Science Focus']} - for example, Columbus studies astronomy to navigate, using constellations like the North Star to guide him. He notes that the Earth's rotation causes the stars to appear in different positions throughout the night, which affects his latitude and longitude measurements.
English Focus: {row['English Focus']} - for example, Columbus uses vocabulary like latitude, longitude, navigation, and exploration to describe his journey, emphasizing the importance of precise measurements and cultural exchange.
Interdisciplinary Theme: {row['Interdisciplinary Theme']} - for example, Columbus's journey demonstrates the theme of exploration, where knowledge guides him through the unknown. He must balance scientific observations with practical navigation skills, showcasing teamwork and leadership.
Moral: {row['Moral']} - for example,Columbus's determination shows that success requires careful planning and hard work, even in the face of uncertainty, highlighting the importance of perseverance and respect for others' cultures.
Use these to illustrate how Columbus applies the chapter's concepts to overcome challenges, demonstrating curiosity and a thirst for adventure.
Data Points and Examples (Paragraph 4):
Include specific data points, statistics, or real-world examples to make the concepts more tangible and engaging. Use these to highlight Columbus's journey and challenges, showcasing his resilience and determination. For example, Columbus's ships travel at an average speed of 5 km/h, covering about 120 km per day. He must manage his resources carefully, as his crew consumes about 180 liters of water daily. This requires efficient use of supplies and strategic planning to ensure they reach their destination safely.
Challenge to Proceed and Conclusion (Paragraph 5):
Present a puzzle or challenge that Columbus must solve to proceed to the next chapter. This should require critical thinking and application of the chapter's concepts, while setting the stage for the next chapter's events. Conclude the story with a sense of accomplishment and anticipation for the next chapter, leaving the reader curious about what lies ahead. For instance, Columbus needs to calculate how long it will take to reach a certain island if he sails at 5 km/h and the distance is 250 km, which involves using multiplication and division.
Challenge:
Develop a multidisciplinary challenge that integrates the math, science, and English focus areas from the chapter. This challenge should require problem-solving skills and critical thinking, encouraging readers to engage deeply with the material. For example, if Columbus has 360 units of food and needs to divide them among his crew using a ratio of 2:3 for sailors to officers, how much food should each group receive? This requires understanding of fractions and ratios.
Questions:
Create 15 questions that assess comprehension and critical thinking related to the chapter's topics. These questions should be styled similarly to academic assessments like NJSLA, STAAR, ISEE, and SSAT, covering a range of skills including problem-solving, data analysis, and thematic understanding. Include multiple-choice, fill-in-the-blank, map-related, punctuation correction, word problems, vocabulary, science, and reasoning questions to ensure a comprehensive assessment. For example:
Multiple Choice: What is the average speed of Columbus's ships?
A) 3 km/h
B) 5 km/h
C) 7 km/h
D) 10 km/h
Fill in the Blanks: Columbus uses _______________________ to navigate during his journey.
Map the Following: Identify the Caribbean Sea on a world map.
Punctuation: Correct the punctuation in the following sentence: "Columbus sailed across the ocean he was very brave"
Word Problem: If Columbus's crew consumes 180 liters of water daily and he has 900 liters available, how many days will the water last?
Vocabulary: What does the word "determined" mean in the context of Columbus's story?
Science: How does the Earth's rotation affect navigation?
Reasoning: Why is it important for Columbus to use astronomy in navigation?
Multiple Choice: What is the main theme of Columbus's journey?
A) Exploration
B) Trade
C) Conquest
D) Settlement
Fill in the Blanks: Columbus's journey demonstrates the importance of _______________________ in overcoming challenges.
Map the Following: Identify the North Star on a star chart.
Punctuation: Correct the punctuation in the following sentence: "Columbus was excited to set sail he had been planning for years"
Word Problem: If Columbus's ship sails for 10 hours at 5 km/h, how far does it travel?
Vocabulary: What does the word "latitude" mean in the context of navigation?
Science: How do wind and ocean currents affect navigation?
"""
    messages = [
        {"role": "system", "content": ""},
        {"role": "user", "content": prompt}
    ]
    response = client.chat.completions.create(
        model="sonar-pro",  # You can change the model if needed
        messages=messages
    )
    return response.choices[0].message.content
```

```python
narratives = []
for index, row in df.iterrows():
    chapter_data = {
        "Chapter": row['Chapter'],
        "Title": row['Title'],
        "Math Focus": row['Math Focus'],
        "Science Focus": row['Science Focus'],
        "English Focus": row['English Focus'],
        "Interdisciplinary Theme": row['Interdisciplinary Theme'],
        "Moral": row['Moral']
    }

    narrative = generate_prompt(chapter_data)
    narratives.append(narrative)  # Append the narrative to the list

# Move the assignment outside the loop to avoid length mismatch
df['Narrative'] = narratives
```
